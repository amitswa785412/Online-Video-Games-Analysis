```python
#check columns with missing data
def assess_missing_data(df):
    
    rec = []
    for column_name in df.columns:
        miss_count = df[column_name].isnull().sum(axis=0)
        miss_percent=miss_count/df.shape[0]
        rec.append([column_name,miss_count,miss_percent*100])
        df_stats = pd.DataFrame.from_records(rec, columns = ['column name', 
                                                             'missing_count','missing_percentage'])  
        df_stats = df_stats.sort_values('missing_percentage',axis=0, ascending = False )
    return df_stats

df_stats = assess_missing_data(vg)
```

```python
#Sorting the boxes in the boxplot in order of ascending average global sales 
order = vg_no_outliers.groupby(['genre']).mean()['global_sales'].sort_values(ascending=True).index.tolist()

#Creating the boxplot
sb.boxplot(data=vg_no_outliers,x='genre',y='global_sales',color=sb.color_palette()[0],order=order);
plt.xlabel('Genre of Video Game')
plt.ylabel('Average Global Sales')
plt.title('Average Global Sales of Different Genres')
plt.xticks(rotation='90');
```

```python
# This shows that indeed some genres like the "Sports and "Shooter" genres are on average more successful (higher sales) than other genres. But perhaps some genres are more stable than others this will be investigated in multivariate exploration.

# A question can be asked which of the specific sales figures is most likely to correlate with the global sales. This can be determined from a scatter plot by identifying which of these features most forms a linear relationship when plotted against global sales. Furthermore this can be checked using pearson's correlation coefficient.
```

```python
graph = sb.PairGrid(data=vg_no_outliers,vars=['na_sales','eu_sales','jp_sales','other_sales','global_sales'])
graph.map_offdiag(plt.scatter,alpha=0.05);
graph.map_diag(plt.hist,bins=15)
```

```python
# Checking with Pearson's correlation coefficient
from scipy.stats import pearsonr
corr,_ = pearsonr(vg_no_outliers['global_sales'],vg_no_outliers['na_sales'])
print("NA sales: " + str(corr))
corr,_ = pearsonr(vg_no_outliers['global_sales'],vg_no_outliers['eu_sales'])
print("EU sales: " + str(corr))
corr,_ = pearsonr(vg_no_outliers['global_sales'],vg_no_outliers['jp_sales'])
print("JP sales: " + str(corr))
corr,_ = pearsonr(vg_no_outliers['global_sales'],vg_no_outliers['other_sales'])
print("other sales: " + str(corr))
```

```python
# This shows that NA sales correlates linearly the most with global sales; this could be due to how big the video game market is in North America.

# Next I would like to investigate whether the video game industry as a whole has an upwards or downwards trend in video game sales.
```

```python
graph = sb.FacetGrid(data=vg_no_outliers[vg_no_outliers['year']>=2005],col='genre',col_wrap=4,sharex=True, sharey=True)
graph.map(sb.pointplot,'year','global_sales', order = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]);
graph.map(plt.xticks,rotation=90);
graph.set_xlabels('Year');
graph.set_ylabels('Average Global Sales\n(in Millions of Copies)');
```

```python
# We can see here that the Adventure genre seems to be very stable over the years with the average sales being around 100,000 while the "Fighting" genre seems to experience ups and downs over the years.

# Another point that I'd like to investigate is the proportion of the different regions from the global sales.
```

```python
#Creating a new dataframe grouped by the years after 2005 with the sum of each of the sales figures features recorded in each year's record
vg_years = vg_no_outliers[['year','na_sales','eu_sales','jp_sales','other_sales','global_sales']][vg_no_outliers['year'] >= 2005]
vg_years = vg_years.groupby(['year']).sum()
```

```python
#Creating new columns to hold the proportions of the sales figures
props_and_sales = [('na_prop','na_sales'),('eu_prop','eu_sales'),('jp_prop','jp_sales'),('other_prop','other_sales')]
for prop,sales in props_and_sales:
    vg_years[prop] = vg_years[sales]/vg_years['global_sales']
vg_years['year']=vg_years.index
vg_years = vg_years[vg_years['year']<2017]
```

```python
vg_years.plot(x='year',y=['na_prop','eu_prop','jp_prop','other_prop'],kind='bar',ylabel='Proprtion of Global Sales')
plt.legend(['NA Sales Proportion','EU Sales Proportion','JP Sales Proportion','Other Sales Proportion']);
plt.xlabel('Year');
plt.title("Different Regions' Proportion of Global Sales Per Year");
```

```python
# Over the years it appears North American sales almost always have the highest proportion however in the recent years this proportion difference is decreasing with other regions getting higher proportions.

# In this section we observed the means of the different genres over the years and discovered that some genres are less stable than others. Additionally, we observed that North American sales over the years have held the most proportion of global sales.
```
